```python
import pandas as pd
import numpy as np 
import random 
from keras.datasets import mnist
import matplotlib.pyplot as plt
from keras import models
from keras import layers
from keras.layers import Dense, Dropout
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
%matplotlib inline

random.seed(1234)
```

```python
# load the dataset 
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
train_images = train_images.reshape((60000, 28 * 28)).astype('float32') / 255
test_images = test_images.reshape((10000, 28 * 28)).astype('float32') / 255
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)

train_images, valid_images, train_labels, valid_labels = \
    train_test_split(train_images, train_labels, test_size = 0.16666, random_state = 1234)
```

```python
network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))
network.add(layers.Dense(512, activation='relu'))
network.add(layers.Dense(512, activation='relu'))
network.add(layers.Dense(512, activation='relu'))
network.add(layers.Dense(10, activation='softmax'))
network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
result = network.fit(train_images, train_labels, 
                     validation_data=(valid_images, valid_labels),
                     epochs=200, batch_size=512)
```

```python
test_loss, test_acc = network.evaluate(test_images, test_labels)
print('test_acc:', test_acc)
```

```python
val_acc = result.history['val_acc']
val_loss = result.history['val_loss']
plt.plot(val_loss)
plt.plot(val_acc)
plt.legend(['validation accurancy', 'validation loss'])
plt.title('The validation set accuracy and loss over epochs')
plt.show()
```

#### implement dropout

```python
model = models.Sequential()
model.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(10, activation='softmax'))
model.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
result_i = model.fit(train_images, train_labels, 
                     validation_data=(valid_images, valid_labels),
                     epochs=200, batch_size=512)
```

```python
val_loss_i = result_i.history['val_loss']
plt.plot(val_loss)
plt.plot(val_loss_i)
plt.legend(['validation loss of original model', 'validation loss of dropout model'])
plt.title('The validation set loss over epochs')
plt.show()
```

The model with drop out performs better than the original model when the number of epochs is large.

#### regularization 

```python
# Penalty: 0.001
from keras import regularizers
model1 = models.Sequential()
model1.add(Dense(512, activation='relu', input_shape=(28 * 28,), kernel_regularizer=regularizers.l1(0.001)))
model1.add(layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l1(0.001)))
model1.add(layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l1(0.001)))
model1.add(layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l1(0.001)))
model1.add(layers.Dense(10, activation='softmax'))
model1.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
result1 = model1.fit(train_images, train_labels, 
                     validation_data=(valid_images, valid_labels),
                     epochs=200, batch_size=512)
```

```python
# Penalty: 0.002
model2 = models.Sequential()
model2.add(Dense(512, activation='relu', input_shape=(28 * 28,), kernel_regularizer=regularizers.l2(0.001)))
model2.add(layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001)))
model2.add(layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001)))
model2.add(layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001)))
model2.add(layers.Dense(10, activation='softmax'))
model2.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
result2 = model2.fit(train_images, train_labels, 
                     validation_data=(valid_images, valid_labels),
                     epochs=200, batch_size=512)
```

```python
val_loss1 = result1.history['val_loss']
val_loss2 = result2.history['val_loss']
plt.plot(val_loss1)
plt.plot(val_loss2)
plt.plot(val_loss)
plt.plot(val_loss_i)
plt.legend(['validation loss of L1 Penalty', 'validation loss of L2 Penalty', 
            'validation loss of original model','validation loss of Model with Dropout'])
plt.title('The validation set loss over epochs')
plt.show()
```

It can be seen that the dropout model has the best performance overall.
